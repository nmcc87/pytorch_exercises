## 2. Simple operations

```python
# Simple numpy-style slicing -> set second row to zero
tensor = torch.ones(4,4)
tensor[:,1] = 0
print(tensor)
```

```python
# Concatenating tensors with cat and stack
tcat = torch.cat(tensors=[tensor,tensor,tensor], dim=2) # dim is the dimension of concatenation; in our case dim=2 means we get a tensor of shape [4,4,12]
print(f"Concatenated shape: {tcat.shape}")

# For stack, dim is the NEW dimension inserted. So in our example we stack 3 tensors, hence tstack will have shape [4,4,3,4]
tstack = torch.stack(tensors=[tensor,tensor,tensor], dim=2) 
print(f"Concatenated shape: {tstack.shape}")
```

```python
# tensor product element wise
tensor.mul(tensor).shape

# matmul product is complicated!

# if both are 1d, returns dot product
t1 = torch.ones(5)
t2 = torch.ones(5)
print(f"Product: {t1.matmul(t2)}")

# if both are 2d, returns matrix product
t1 = torch.ones(5,3)
t2 = torch.ones(3,4)
print(f"Product: {t1.matmul(t2)}")
print(f"Product shape: {t1.matmul(t2).shape}")

# 1d times 2d returns a vector
t1 = torch.ones(3)
t2 = torch.ones(3,4)
print(f"Product: {t1.matmul(t2)}")
print(f"Product shape: {t1.matmul(t2).shape}")

# 2d times 1d returns a vector
t1 = torch.ones(3,4)
t2 = torch.ones(4)
print(f"Product: {t1.matmul(t2)}")
print(f"Product shape: {t1.matmul(t2).shape}")

# Now the fun: if at least one tensor has N>2 dims, the multiplication returns a batch matmul
# Batch dims are the first N-2 (which are broadcast)
# Broadcasting: from right to left, dims are compared and are compatible if either dim1==dim2 or one of the 2 is 1
# Then the 1-dim are treated as vectors
# Last N-2 are multiplied
t1 = torch.ones(7,2,5,6)
t2 = torch.ones(7,2,6,4)
print(f"Product shape: {t1.matmul(t2).shape}")

# with 1d broadcasting
t1 = torch.ones(7,2,5,6)
t2 = torch.ones(1,1,6,4)
print(f"Product shape: {t1.matmul(t2).shape}")

```

## 3. numpy vs torch

```python
# numpy arrays and torch tensors share memory allocation, so changes in one are reflected in the other
n = np.ones(5)
t = torch.from_numpy(n)
print(f"torch tensor pre addition: {t}")
np.add(n, 1, out=n)
print(f"torch tensor post addition: {t}")
```

```python
# but remember: variable renaming reallocates memory!
n = np.ones(5)
t = torch.from_numpy(n)
print(f"torch tensor pre addition: {t}")
n = np.add(n, 1)
print(f"torch tensor post addition: {t}")
```
