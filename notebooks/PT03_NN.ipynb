# Neural Networks 101

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
```

```python
class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()

        # Define simple architecture for CNN
        # Layer 1: 1 input, 5 output, 5x5 kernel
        self.conv1 = nn.Conv2d(in_channels=1, out_channels=6, kernel_size=5)
        # Layer 2: 6 input, 16 output, 5x5 kernel
        self.conv1 = nn.Conv2d(in_channels=6, out_channels=16, kernel_size=5)

        # Add an affine operation y = Wx+b, maps the 16*5*5 output to a 10d vector
        self.fc1 = nn.Linear(in_features=16*5*5,out_features=120)
        self.fc2 = nn.Linear(in_features=120,out_features=84)
        self.fc3 = nn.Linear(in_features=84,out_features=10)

    def forward(self, input):
        """
        Forward pass implementation
        """
        # Step 1: 1 input channel, 6 output channels, 5x5 kernel, RELU activation
        # Input: Batch of N images, size 32x32
        # Output: (N, 6, 28, 28) tensor (28 = 32-5+1, kernel is 5x5)
        c1 = F.relu(self.conv1(input))
        # Pooling, slice dim by 2
        # Output: (N, 6, 14, 14)
        s2 = F.max_pool2d(c1, (2,2))
        # Step 2, 6 input channels, 16 output, 5x5 kernel, RELU activation
        # Output: (N, 16, 10, 10)
        c3 = F.relu(self.conv2(s2))
        # Pooling
        s4 = F.max_pool2d(c3, (2,2))
        # Flatten: from (N, 16, 5, 5) to batch-flat (N, 400)
        s4 = torch.flatten(s4, 1)
        # first affine reduction 400 -> 120 + RELU
        f5 = F.relu(self.fc1(s4))
        # second affine reduction 120 -> 84 + RELU
        f6 = F.relu(self.fc2(f5))
        # third affine reduction 84-> 10, no RELU
        output = self.fc3(f6)
        return output
        
```
